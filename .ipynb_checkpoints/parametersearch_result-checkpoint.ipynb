# Hyperparameter Analysis

```python
import os
import glob
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tiny_hanabi.game.payoff_matrices import OPTIMAL_RETURNS
# --- VISUAL SETUP ---
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 6]

# --- CONFIGURATION ---
INPUT_RESULTS_DIR = "Results/Baselines" 
FINAL_OUTPUT_BASE = "Results/Baselines" 
GAMES = ["A", "B", "C", "D", "E", "F"]
```

```python
def load_hyperparameter_results(agent_name):
    """
    Reads all search results for a given agent. Returns summary DataFrame.
    """
    folder_name = agent_name.replace(" ", "_")
    search_dir = os.path.join(INPUT_RESULTS_DIR, folder_name, "Search Results")
    
    if not os.path.exists(search_dir):
        print(f"❌ Directory not found: {search_dir}")
        return pd.DataFrame()

    csv_files = glob.glob(os.path.join(search_dir, "*_results.csv"))
    summary_data = []

    for csv_path in csv_files:
        try:
            file_name = os.path.basename(csv_path)
            idx = int(file_name.split("_")[0])
            
            # Load Data & Params
            df = pd.read_csv(csv_path)
            json_path = os.path.join(search_dir, f"{idx}_params.json")
            with open(json_path, 'r') as f:
                params = json.load(f)
            
            # Calculate Final Reward (Last Episode Average)
            reward_cols = [c for c in df.columns if "reward" in c]
            final_reward = df.iloc[-1][reward_cols].mean()
            
            entry = {"id": idx, "final_reward": final_reward, **params}
            summary_data.append(entry)
        except Exception as e:
            pass # Skip corrupted files

    if not summary_data:
        return pd.DataFrame()

    return pd.DataFrame(summary_data).sort_values("id").reset_index(drop=True)
```

```python
def plot_search_overview(df, agent_name):
    """
    Plots 1. Scatter of ID vs Reward, 2. Histogram of Rewards (binwidth 0.2).
    """
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))
    
    # 1. Scatter Plot
    sns.scatterplot(
        data=df, x="id", y="final_reward", hue="final_reward", 
        palette="viridis", s=100, edgecolor="black", ax=axes[0], legend=False
    )
    axes[0].set_title(f"{agent_name}: Search Distribution", fontsize=14)
    axes[0].set_xlabel("Parameter Set ID")
    axes[0].set_ylabel("Average Final Reward")
    axes[0].grid(True, alpha=0.5)

    # 2. Histogram
    sns.histplot(
        data=df, x="final_reward", binwidth=0.05, 
        kde=True, color="teal", ax=axes[1]
    )
    axes[1].set_title(f"{agent_name}: Performance Frequency (Bin=0.05)", fontsize=14)
    axes[1].set_xlabel("Final Reward")
    axes[1].set_ylabel("Count")
    axes[1].grid(True, alpha=0.5)
    
    plt.tight_layout()
    plt.show()
```

```python
def visualize_agent_performance(summary_df, agent_name, window_size=50):
    """
    Generates a separate figure for each of the Top 5 Parameter Sets.
    Includes explicit Y-axis scaling to ensure the Optimal Line is visible.
    """
    #summary_df = load_hyperparameter_results(agent_name)
    
    if summary_df.empty:
        print(f"No data for {agent_name}")
        return

    top_5_df = summary_df.sort_values(by="final_reward", ascending=False).head(5).reset_index(drop=True)
    
    if len(top_5_df) == 0:
        print("No parameter sets found.")
        return

    # Calculate Global Average Optimal Return
    global_avg_optimal = np.mean([OPTIMAL_RETURNS[g] for g in GAMES])

    for index, param_row in top_5_df.iterrows():
        
        param_id = int(param_row['id'])
        final_rew = param_row['final_reward']
        
        folder_name = agent_name.replace(" ", "_")
        csv_path = os.path.join(INPUT_RESULTS_DIR, folder_name, "Search Results", f"{param_id}_results.csv")
        
        if not os.path.exists(csv_path):
            continue
            
        ts_df = pd.read_csv(csv_path)
        
        # --- PRINT SUMMARY ---
        print("\n" + "="*80)
        print(f"🏆 RANK {index+1} | ID: {param_id} | Avg Reward: {final_rew:.4f} / {global_avg_optimal:.4f}")
        print("-" * 40)
        excluded = ['id', 'final_reward']
        print(json.dumps({k:v for k,v in param_row.items() if k not in excluded}, indent=4))
        print("="*80 + "\n")

        # --- PLOT ---
        fig, axes = plt.subplots(nrows=1, ncols=7, figsize=(28, 5)) # Removed sharey=True to allow individual scaling
        
        # Helper to plot data + optimal line
        def plot_on_axis(ax, data, label, color, optimal_value):
            # 1. Calc Moving Average
            if len(data) < window_size:
                ma = data.expanding().mean()
            else:
                ma = data.rolling(window=window_size, min_periods=1).mean()
            
            # 2. Plot Data
            ax.plot(data, color='gray', alpha=0.15, linewidth=1)
            ax.plot(ma, color=color, linewidth=2.5, label='Agent')
            
            # 3. Plot Optimal Line & Fix Scaling
            max_data_val = data.max() if not data.empty else 0.0
            
            if optimal_value is not None:
                ax.axhline(y=optimal_value, color='green', linestyle='--', linewidth=2, alpha=0.8, label='Optimal')
                
                # --- FIX: Force Y-axis to include the optimal value ---
                # We take the max of the data OR the optimal value, plus 10% padding
                upper_limit = max(max_data_val, optimal_value) * 1.1
                ax.set_ylim(bottom=0, top=upper_limit)
            else:
                ax.set_ylim(bottom=0, top=max_data_val * 1.1)

            ax.set_title(label, fontsize=14, fontweight='bold')
            ax.grid(True, linestyle='--', alpha=0.6)
            ax.set_xlabel("Episode")

        # 4. Plot Games A-F
        reward_cols = []
        for col_idx, game in enumerate(GAMES):
            ax = axes[col_idx]
            reward_col = f"reward_{game}"
            reward_cols.append(reward_col)
            
            opt_val = OPTIMAL_RETURNS.get(game, None)
            
            if reward_col in ts_df.columns:
                plot_on_axis(ax, ts_df[reward_col], label=f"Game {game}", color='tab:blue', optimal_value=opt_val)
            else:
                ax.text(0.5, 0.5, "No Data", ha='center', va='center')

        # 5. Plot Average
        ax_avg = axes[6]
        valid_cols = [c for c in reward_cols if c in ts_df.columns]
        if valid_cols:
            avg_series = ts_df[valid_cols].mean(axis=1)
            plot_on_axis(ax_avg, avg_series, label="AVERAGE", color='darkred', optimal_value=global_avg_optimal)
        else:
            ax_avg.text(0.5, 0.5, "No Data", ha='center', va='center')

        # 6. Finalize
        fig.suptitle(f"Training Trajectories (Rank {index+1} - ID {param_id})", fontsize=16, y=1.05)
        axes[0].set_ylabel("Reward", fontsize=12)
        
        # Legend on the last plot
        handles, labels = axes[6].get_legend_handles_labels()
        if handles:
            axes[6].legend(handles, labels, loc='lower right', fontsize='small')

        plt.tight_layout()
        plt.show()
```

```python
def save_best_params(agent_name, params_dict):
    folder_name = agent_name.replace(" ", "_")
    output_dir = os.path.join(FINAL_OUTPUT_BASE, folder_name)
    os.makedirs(output_dir, exist_ok=True)
    file_path = os.path.join(output_dir, "best_params.json")
    with open(file_path, 'w') as f:
        json.dump(params_dict, f, indent=4)
    print(f"✅ Params saved to: {file_path}")
```

```python
def analyze_agent(agent_name, window_size=50):
    df = load_hyperparameter_results(agent_name)
    if df.empty:
        print(f"No data for {agent_name}")
        return
    
    # 1. Overview
    plot_search_overview(df, agent_name)
    
    # 2. Detailed Top 5
    visualize_agent_performance(df, agent_name, window_size)
```

### Model Free - Independent Learners

```python
AGENT_NAME = "Independent_RL_Agent"
analyze_agent(AGENT_NAME, window_size=50)
```
