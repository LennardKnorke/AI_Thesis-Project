## Results

```python
import os
import glob
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tiny_hanabi.game.payoff_matrices import OPTIMAL_RETURNS
```

```python
# --- VISUAL SETUP ---
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 6]

# --- CONFIGURATION ---
INPUT_RESULTS_DIR = "HyperSearchResults" 
FINAL_OUTPUT_BASE = "Results" 
GAMES = ["A", "B", "C", "D", "E", "F"]
AGENTS = ["Independent_RL_Agent", "Independent_VI_Agent", "VDN_RL_Agents", "VDN_VI_Agents"]
```

```python
def load_df(agent_name : str):
    """
    Read search results for the given agent
    """
    folder_name = agent_name.replace(" ", "_")
    search_dir = os.path.join(INPUT_RESULTS_DIR, folder_name)
    
    if not os.path.exists(search_dir):
        print(f"❌ Directory not found: {search_dir}")
        return pd.DataFrame(), {}
        
    results_path = os.path.join(search_dir, "final_results.csv")
    if not os.path.exists(results_path):
        print(f"Results not found: {results_path}")
        return pd.DataFrame(), {}
        
    params_path = os.path.join(search_dir, "best_params.json")
    if not os.path.exists(params_path):
        print(f"Results not found: {params_path}")
        return pd.DataFrame(), {}

    df = pd.read_csv(results_path)

    with open(params_path) as json_data:
        params = json.load(json_data)
    return df, params
```

#### Model-Free RL

```python
baselines_df = {}
baselines_params = {}
for agent in AGENTS:
    df, params = load_df(agent)
    baselines_df[agent] = df
    baselines_params[agent] = params
```
